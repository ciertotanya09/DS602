## Data Definitions

- `department`: Department person works(worked).

- `gender`: Gender of individual.  
- `seniority_level`: Seniority level of individual.

- `married_or_single`: Marital status of person.

- `years_at_company`: Number of years the individual has been at the company, regardless of position.  
- `years_in_current_job`: Number of years the individual has been in their current position.  
- `years_since_last_promotion`: Years since the person had their last promotion.  
- `years_with_current_supervisor`: Years the person has had their current supervisor.

Encoding the frequency of travel with 1 for non travel, 2 for the employees who travel rarely and 3 for the employees who travel frequently

```python
# creating tenure-to-promotion ratio
df['tenure_to_promotion'] = df['years_at_company'] / (df['years_since_last_promotion'] + 1)

# create work-life balance indicator
df['work_life_balance'] = (
    df['commuting_distance'] +  
    df['frequency_of_travel'].map({'Non-Travel': 1, 'Travel_Rarely': 2, 'Travel_Frequently': 3}).fillna(0) +  
    (5 - df['satisfaction_with_job']) +  
    df['years_with_current_supervisor']  
)

# creating job satisfaction delta
df['satisfaction_delta'] = df['satisfaction_with_job'] - df['satisfaction_with_environment']

# creating binary feature indicating long wait time for promotion (e.g., more than 3 years)
df['long_promotion_wait'] = df['years_since_last_promotion'].apply(lambda x: 1 if x > 3 else 0)
```

```python
# long_promotion_wait vs left_company
plt.figure(figsize=(10, 6))
sns.countplot(x='long_promotion_wait', hue='left_company', data=df)
plt.title('Impact of long promotion wait time on leaving company')
plt.show()
```

> More employees overall had normal/shorter wait times for promotion. Among those with normal/shorter wait times, slightly more left than stayed. Among those with long wait times, more stayed than left

```python
# compute the correlation matrix focusing on numeric features
plt.figure(figsize=(12,8))
numeric_cols = ['age', 'commuting_distance', 'education', 'satisfaction_with_environment', 'seniority_level',
                'satisfaction_with_job', 'last_raise_pct', 'last_performance_rating', 
                'total_years_working', 'years_at_company', 'years_in_current_job',
                'years_since_last_promotion', 'years_with_current_supervisor','tenure_to_promotion','work_life_balance',
                'satisfaction_delta','long_promotion_wait','left_company']
corr_matrix = df[numeric_cols].corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation matrix of numeric features')
plt.show()
```

# Question 4
- Split the data into training and test sets.  
- Use 20% of the data for test and a random state of 124.  

```python
# insert code here
from sklearn.model_selection import train_test_split

X = df_numerical.drop('left_company', axis=1)
y = df_numerical['left_company']

# split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=124)

# checking the shape of the resulting datasets
print(f'Training set size: {X_train.shape[0]}')
print(f'Test set size: {X_test.shape[0]}')
```

# Question 5
- Create a pipeline to process the numerical data.  
- Create a pipeline to process the categorical data.  

Verify each pipeline contains the columns you would expect using a fit_transform on the training data, i.e., print the shapes of the fit_transforms for each pipeline.

```python
# insert code here
# importing the necessary libraries
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

# defining the numerical features except for the target variable
numerical_features = df_numerical.select_dtypes(include=['int64', 'float64']).columns.tolist()
numerical_features.remove('left_company')

# creating a pipeline for numerical data
numerical_pipeline = Pipeline(steps=[
    ('scaler', StandardScaler())
])

# fit and transform the training data
X_train_transformed = numerical_pipeline.fit_transform(X_train[numerical_features])

# verify the shape of the transformed data
print(f'Shape of transformed numerical data: {X_train_transformed.shape}')
```

# Question 6
- Create a pipeline that combines the pre-processing and implements a logistic regression model.  
- Print the accuracy on the training set and the test set.
- Do you have any concerns of overfitting based on the differences between the two accuracy scores?

```python
# insert code here
# importing the necessary libraries
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# creating a pipeline that combines preprocessing and logistic regression
model_pipeline = Pipeline(steps=[
    ('preprocessor', numerical_pipeline),  # numerical preprocessing
    ('classifier', LogisticRegression())   # logistic regression model
])

# fit the pipeline to the training data
model_pipeline.fit(X_train, y_train)

# predict on the training set and test set
y_train_pred = model_pipeline.predict(X_train)
y_test_pred = model_pipeline.predict(X_test)

# calculate accuracy on the training set and test set
train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)

# print the accuracy scores
print(f'Training Accuracy: {train_accuracy:.2f}')
print(f'Test Accuracy: {test_accuracy:.2f}')
```

# Question 7
What would you recommend as potential next steps for continuing to develop and evaluate a model?

```python
# importing the necessary libraries
from sklearn.metrics import confusion_matrix

# make predictions on the test set
y_test_pred = model_pipeline.predict(X_test)

# calculate the confusion matrix
cm = confusion_matrix(y_test, y_test_pred)

# creating a DataFrame for better visualization
cm_df = pd.DataFrame(cm, index=['Not Left', 'Left'], columns=['Not Left', 'Left'])

# show the confusion matrix
plt.figure(figsize=(4,2))
sns.heatmap(cm_df, annot=True, fmt='d', cmap='Blues', linewidths=0.5)
plt.title('Confusion matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
# importing the necessary libraries
from sklearn.metrics import roc_curve, roc_auc_score

# getting the predicted probabilities (for the positive class, i.e., left_company = 1)
y_train_prob = model_pipeline.predict_proba(X_train)[:, 1]
y_test_prob = model_pipeline.predict_proba(X_test)[:, 1]

# compute ROC curve
fpr_train, tpr_train, _ = roc_curve(y_train, y_train_prob)
fpr_test, tpr_test, _ = roc_curve(y_test, y_test_prob)

# plotting the ROC curve for both training and test sets
plt.figure(figsize=(10, 6))
plt.plot(fpr_train, tpr_train, label="Training ROC Curve")
plt.plot(fpr_test, tpr_test, label="Test ROC Curve")
plt.plot([0, 1], [0, 1], color='black', linestyle='-')
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('ROC Curve')
plt.legend()
plt.show()

# calculate the AUC score
train_auc = roc_auc_score(y_train, y_train_prob)
test_auc = roc_auc_score(y_test, y_test_prob)

print(f"Training AUC: {train_auc}")
print(f"Test AUC: {test_auc}")
```

> The ROC curve shows that the model is moderately effective at distinguishing between employees who left and those who stayed. The training AUC is 0.78, indicating decent performance on the training data, while the test AUC is 0.71, showing that the model generalizes reasonably well but might still benefit from improvements. The gap suggests a slight overfitting, but it's not drastic.
